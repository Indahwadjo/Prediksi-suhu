```python
from graphviz import Digraph
from IPython.display import Image, display

class TreeNode:
    def __init__(self, feature=None, threshold=None, left=None, right=None, prediction=None):
        self.feature = feature
        self.threshold = threshold
        self.left = left
        self.right = right
        self.prediction = prediction

    def is_leaf(self):
        return self.prediction is not None


def build_manual_tree():
    cerah = TreeNode(prediction="Cerah")
    cerah_berawan = TreeNode(prediction="Cerah Berawan")
    berawan = TreeNode(prediction="Berawan")
    hujan = TreeNode(prediction="Hujan")

    root = TreeNode(
        feature="Suhu", threshold=24,
        left=hujan,
        right=TreeNode(
            feature="Suhu", threshold=26,
            left=TreeNode(
                feature="Kelembaban", threshold=75,
                left=TreeNode(
                    feature="Kecepatan Angin", threshold=21.6,
                    left=berawan,
                    right=hujan
                ),
                right=hujan
            ),
            right=TreeNode(
                feature="Suhu", threshold=28,
                left=TreeNode(
                    feature="Kelembaban", threshold=75,
                    left=TreeNode(
                        feature="Kecepatan Angin", threshold=10.8,
                        left=cerah_berawan,
                        right=hujan
                    ),
                    right=hujan
                ),
                right=TreeNode(
                    feature="Kelembaban", threshold=65,
                    left=TreeNode(
                        feature="Kecepatan Angin", threshold=5.4,
                        left=cerah,
                        right=cerah_berawan
                    ),
                    right=cerah_berawan
                )
            )
        )
    )
    return root

def visualisasi_pohon_keputusan(node):
    dot = Digraph()
    dot.attr('node', shape='box', style='filled', fontname='Arial')

    node_counter = [0]

    def add_nodes_edges(current_node, parent_id=None, edge_label=""):
        node_id = f"node{node_counter[0]}"
        node_counter[0] += 1

        if current_node.is_leaf():
            label = f"Prediksi:\n{current_node.prediction}"
            dot.node(node_id, label, fillcolor='lightblue')
        else:
            label = f"{current_node.feature}\n<= {current_node.threshold}"
            dot.node(node_id, label, fillcolor='lightgreen')

        if parent_id is not None:
            dot.edge(parent_id, node_id, label=edge_label)

        if not current_node.is_leaf():
            add_nodes_edges(current_node.left, node_id, "Ya")
            add_nodes_edges(current_node.right, node_id, "Tidak")

    add_nodes_edges(node)

    file_path = "pohon_keputusan_cuaca"
    dot.render(file_path, format='png', cleanup=True)
    display(Image(filename=f"{file_path}.png"))

root = build_manual_tree()
visualisasi_pohon_keputusan(root)
```
